```python
df_canada.sort_values(by=["total"], ascending=False).head()
```

```python
top5_countries = df_canada.sort_values(by=["total"], ascending=False).head()
top5_countries = top5_countries[list_years]
top5_countries = top5_countries.T
top5_countries
```

```python
top5_countries.plot(
    kind="area",
    figsize=(20, 10)
)
plt.title("Top 5 Immigrant's Origin Countries")
plt.ylabel("Num of Imigrants")
plt.xlabel("Years")
plt.show()
```

```python
top5_countries.plot(
    kind="area",
    figsize=(20, 10),
    stacked=False
)
plt.title("Top 5 Immigrant's Origin Countries")
plt.ylabel("Num of Imigrants")
plt.xlabel("Years")
plt.show()
```

```python
top15_countries = df_canada.sort_values(by="total", ascending=False).head(15)
top15_countries = top15_countries["total"]
top15_countries.head()
```

```python
continents = df_canada.groupby("continent").sum()
continents
```

```python
explode_list = [.1, 0, 0, 0, .1, .1]
continents["total"].plot(
    kind="pie",
    figsize=(15, 6),
    autopct="%.1f%%",
    startangle=90,
#   shadow=True,
    labels=None,
    pctdistance=1.05,
    explode=explode_list
)
plt.title("Total Immigrants by Continents")
plt.axis("equal")
plt.legend(labels=continents.index, loc="upper left")
plt.show()
```
